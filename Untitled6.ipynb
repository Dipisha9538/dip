Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-Processing

```python
# loading the data from csv file to apandas dataframe
movies_data = pd.read_csv('/content/netflix_titles.csv')
```

```python
# printing the first 5 rows of the dataframe
movies_data.head()
```

```python
# Rows and column in data frame
movies_data.shape
```

```python
# selecting features for movie recommendation
selected_features = ['type','title', 'director', 'country', 'description' ]
print(selected_features)
```

```python
#replacing the null values with null string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combining all the 5 selected features
combined_features = movies_data['type']+' '+movies_data['title']+' '+movies_data['director']+' '+movies_data['country']+' '+movies_data['description']
```

```python
print(combined_features)
```

```python
#coverting the text data to features vector
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from the user
movie_name = input('Enter Your favourite TV Show or movie name:')
```

```python
#creating all the names of TV Show and movie
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for movie name or TV Show given by user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the show_id of the movie with title
show_id_of_the_movie = movies_data[movies_data.title == close_match]['show_id'].values[0]
print(show_id_of_the_movie)
```

```python
#getting show_id of similar movies
similarity_score = list(enumerate(similarity[show_id_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies and TV Show based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the show_id
print('Movies or TV Show suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  show_id = movie[0]
  title_from_show_id = movies_data[movies_data.show_id==show_id]['title'].values[0]
  if(i<21):
    print(i, '.', title_from_show_id)
    i+=1
```

```python
movie_name = input('Enter Your favourite movie name:')
list_of_all_titles = movies_data['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = find_close_match[0]
show_id_of_the_movie = movies_data[movies_data.title == close_match]['show_id'].values[0]
similarity_score = list(enumerate(similarity[show_id_of_the_movie]))
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  show_id = movie[0]
  title_from_show_id = movies_data[movies_data.show_id==show_id]['title'].values[0]
  if(i<35):
    print(i, '.', title_from_show_id)
    i+=1
```
